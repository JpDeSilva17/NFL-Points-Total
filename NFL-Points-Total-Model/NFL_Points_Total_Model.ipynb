## Background and Prior Work

**Early Statistical Models**

Predicting the **points total** in NFL games has become a key area of sports analytics, particularly in betting markets. Unlike point spreads, which focus on relative performance between teams, total points forecasting requires capturing offensive efficiency, defensive strength, and game tempo while accounting for external factors such as weather.

Early Statistical Models
Initial research into NFL scoring distributions established fundamental principles for total points prediction. Stern (1991) showed that NFL game scores roughly follow a normal distribution, reinforcing the efficiency of betting markets.<a name="cite_ref-1"></a><sup>1</sup> Glickman & Stern (1998) introduced state-space models, which dynamically adjusted offensive and defensive team ratings over time, improving score prediction accuracy compared to static methods.<a name="cite_ref-2"></a><sup>2</sup>

Baker & McHale (2013) advanced this approach with a point process model, which analyzed scoring events in real-time rather than relying on game-level aggregates. Their study demonstrated that their model’s predicted totals closely aligned with sportsbooks’ closing lines, highlighting the difficulty of gaining a consistent predictive edge over betting markets.<a name="cite_ref-3"></a><sup>3</sup>

Simultaneously, external factors were identified as key contributors to total points variance. Borghesi (2008) found that sportsbooks systematically underestimated the impact of extreme weather on low-scoring games, presenting an inefficiency that data-driven models could exploit.<a name="cite_ref-4"></a><sup>4</sup>

**Machine Learning and Advanced Metrics**

Modern approaches to total points prediction leverage machine learning models and advanced football analytics. Metrics such as Expected Points Added (EPA), turnover differential, red zone efficiency, and pace of play have significantly enhanced predictive accuracy.

Lock & Nettleton (2014) applied Random Forest models to estimate in-game win probabilities, demonstrating the power of machine learning to capture complex interactions between game-state variables.<a name="cite_ref-5"></a><sup>5</sup> More recent research indicates that ensemble models (Random Forest, Gradient Boosting) consistently outperform linear regression in forecasting points totals, as they better handle non-linearity and interactions between team performance metrics.<a name="cite_ref-6"></a><sup>6</sup>

One challenge for machine learning models has been capturing game flow and strategic adjustments. Traditional models assume independent scoring events, but real-world factors—such as teams shifting to conservative play-calling in late-game blowouts—affect total points outcomes. This has led to the incorporation of situational modeling, where AI systems analyze game conditions to adjust predictions dynamically.

**Industry Benchmarks and Challenges**

Sportsbooks establish totals using power-rating models that factor in historical performance, team strengths, injuries, and betting patterns. These lines continuously adjust in response to new information and betting activity, making them dynamic benchmarks that machine learning models must compete against.

Despite advances in machine learning, sportsbooks' closing totals remain highly accurate, as they reflect both quantitative models and real-time market sentiment. Studies have found that machine learning models typically match but rarely exceed the accuracy of sportsbook lines, underscoring the difficulty of extracting additional predictive value from public data.<a name="cite_ref-7"></a><sup>7</sup>

## nfl-data-py: aggregated play-by-play game-level data 

````python
import nfl_data_py as nfl
import pandas as pd

# play_by_play_variables = nfl.see_pbp_cols()
# print(play_by_play_variables)
````

````python
# Import the play-by-play data
years = [2021, 2022, 2023, 2024]

columns = [
    'game_id', 'season', 'week', 'yards_gained', 'epa', 'weather',
    'total_home_score', 'total_away_score', 'score_differential',
    'posteam', 'defteam', 'home_team', 'away_team', 'total_home_rush_epa', 
    'total_away_rush_epa', 'drive_time_of_possession', 'sack', 'interception', 
    'fumble_forced', 'rushing_yards', 'receiving_yards', 'third_down_converted', 'third_down_failed'
]



pbp_data = nfl.import_pbp_data(years=years, columns=columns, downcast=True)

# Show the first few rows of the raw data
pbp_data.head()
````

The play-by-play data for each game is raw and needs preparation before it can be modeled for the use case. To start we can see that the data type for ```drive_time_of_possession``` is an object. We need to convert this column to integers in order to have each teams' drive time of possession be represented in seconds as continuous numbers.

````python
# Convert 'drive_time_of_possession' from minutes:seconds format to total seconds
# This makes it easier to perform numerical operations later.
pbp_data['drive_time_of_possession'] = (
    pbp_data['drive_time_of_possession']
    .fillna('0:00')  # Fill missing values with "0:00"
    .apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))  # Convert to seconds
)


pbp_data[['game_id','posteam', 'drive_time_of_possession']].head(10)
````

We have identified where each new drive begins. This shows that the first row of every drive will alternate the team for the ```posteam``` column. After the first row, the ```drive_time_of_possession``` and the ```posteam``` ID repeat for the following rows for the duration of that ongoing drive. This will remain the same until a new drive starts where the ```posteam``` value will then shift back to the other team and a new value for ```drive_time_of_possession``` will repeat. We will flag each time this happens to correctly identify when the ball changes possession.

````python
# Flag rows where a new drive begins
pbp_data['new_drive'] = (
    (pbp_data['drive_time_of_possession'] != 0) &
    (pbp_data['drive_time_of_possession'] != pbp_data['drive_time_of_possession'].shift()) &
    (pbp_data['posteam'] != pbp_data['posteam'].shift())
)

# Calculate time of possession based on flagged new drives
pbp_data['time_of_possession'] = pbp_data['drive_time_of_possession'].where(pbp_data['new_drive'], 0)

pbp_data[['game_id', 'new_drive', 'time_of_possession']].head(10)
````

The column ```new_drive``` will have the value ```True``` and the corresponding time of possession for that drive. All remainding rows for that drive will have the value ```False```. Our data is now ready to be separated into home and away team metrics. This will provide a clear distinction for when each team is on offense or defense. We will do this by identifying which team has possession of the ball for any given play.

````python
# Split data into possessions for home and away teams
home_team_ball = pbp_data[pbp_data['posteam'] == pbp_data['home_team']]

# Display samples of the split data
home_team_ball.head(5)
````

````python
# Split data into possessions for home and away teams
away_team_ball = pbp_data[pbp_data['posteam'] == pbp_data['away_team']]

# Display samples of the split data
away_team_ball.head(5)
````

Now that we have split the plays into home and away possessions, we can aggregate the data for each game. This will ensure that each row in the resulting datasets represents a team's performance for one full game, rather than each play.

````python
# Aggregate performance metrics for home teams
home_data = home_team_ball.groupby('game_id').agg(
    home_total_yards=('yards_gained', 'sum'),
    home_total_epa=('epa', 'sum'),
    home_time_of_possession=('time_of_possession', 'sum'),
    home_rushing_yards=('rushing_yards', 'sum'),
    home_receiving_yards=('receiving_yards', 'sum'),
    home_third_down_converted=('third_down_converted', 'sum'),
    home_third_down_failed=('third_down_failed', 'sum'),
    home_time_to_throw=('time_to_throw', lambda x: x.dropna().mean()),
    away_pressures=('was_pressure', 'sum'),
    away_sacks=('sack', 'sum'),
    away_interceptions=('interception', 'sum'),
    away_fumbles_forced=('fumble_forced', 'sum'),
    home_final_score=('total_home_score', 'last')
).reset_index()

# Aggregate performance metrics for away teams
away_data = away_team_ball.groupby('game_id').agg(
    away_total_yards=('yards_gained', 'sum'),
    away_total_epa=('epa', 'sum'),
    away_time_of_possession=('time_of_possession', 'sum'),
    away_rushing_yards=('rushing_yards', 'sum'),
    away_receiving_yards=('receiving_yards', 'sum'),
    away_third_down_converted=('third_down_converted', 'sum'),
    away_third_down_failed=('third_down_failed', 'sum'),
    away_time_to_throw=('time_to_throw', lambda x: x.dropna().mean()),
    home_pressures=('was_pressure', 'sum'),
    home_sacks=('sack', 'sum'),
    home_interceptions=('interception', 'sum'),
    home_fumbles_forced=('fumble_forced', 'sum'),
    away_final_score=('total_away_score', 'last')
).reset_index()
````

Now, we can calculate key features such as turnover and sack differentials to quantify game dynamics. These differentials help assess which team had a greater impact in forcing turnovers or pressuring the quarterback. Additionally, we can analyze efficiency metrics like third-down conversion rates, explosive play frequency, and average time to throw to gain deeper insights into offensive and defensive performance. By integrating these features, we can better understand the factors that influence game outcomes and team effectiveness.

````python
# Calculate turnover and sack differentials for home and away teams
home_data['home_turnovers'] = away_data['home_interceptions'] + away_data['home_fumbles_forced']
away_data['away_turnovers'] = home_data['away_interceptions'] + home_data['away_fumbles_forced']

home_data['home_turnover_differential'] = home_data['home_turnovers'] - away_data['away_turnovers']
away_data['away_turnover_differential'] = away_data['away_turnovers'] - home_data['home_turnovers']

home_data['home_sack_differential'] = away_data['home_sacks'] - home_data['away_sacks']
away_data['away_sack_differential'] = home_data['away_sacks'] - away_data['home_sacks']

home_data['home_third_down_%'] = home_data['home_third_down_converted'] / (home_data['home_third_down_converted'] + home_data['home_third_down_failed'])
away_data['away_third_down_%'] = away_data['away_third_down_converted'] / (away_data['away_third_down_converted'] + away_data['away_third_down_failed'])
````

It's time to merge our home and away datasets to complete each game. We will aggregate defensive data, include contextual information and fix any inconsistencies.

````python
# Merge home and away team data into a single dataset
game_data = pd.merge(home_data, away_data, on='game_id')

# Aggregate defensive metrics for home and away teams
home_defense_data = pbp_data[pbp_data['defteam'] == pbp_data['home_team']].groupby('game_id').agg(
    home_defense_epa_allowed=('epa', 'sum'),
    home_receiving_yards_allowed=('receiving_yards', 'sum'),
    home_rushing_yards_allowed=('rushing_yards', 'sum')
).reset_index()

away_defense_data = pbp_data[pbp_data['defteam'] == pbp_data['away_team']].groupby('game_id').agg(
    away_defense_epa_allowed=('epa', 'sum'),
    away_receiving_yards_allowed=('receiving_yards', 'sum'),
    away_rushing_yards_allowed=('rushing_yards', 'sum')
).reset_index()

# Merge defensive metrics into the main game data
game_data = pd.merge(game_data, home_defense_data, on='game_id')
game_data = pd.merge(game_data, away_defense_data, on='game_id')

# Add contextual information (season, week, weather) and clean team names
game_context = pbp_data[['game_id', 'season', 'week', 'home_team', 'away_team', 'weather']].drop_duplicates()
game_data = pd.merge(game_data, game_context, on='game_id')

# Standardize inconsistent team names (e.g., "LA" -> "LAR")
game_data.loc[game_data['away_team'] == 'LA', 'away_team'] = 'LAR'
game_data.loc[game_data['home_team'] == 'LA', 'home_team'] = 'LAR'

# Display the final dataset unifying all previous ideas
#print(game_data.shape)
#print(game_data.dtypes)
game_data.head()
````

We can see now that after all the changes have come into effect the data not only is more visually appealing but much more apt for analysis.

### How Accurate Are Sportsbooks At Predicting Over/Unders?

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure numeric types
model_game_data['game_over_under'] = pd.to_numeric(model_game_data['game_over_under'], errors='coerce')
model_game_data['total'] = pd.to_numeric(model_game_data['total'], errors='coerce')

# Drop NaN values
model_game_data = model_game_data.dropna(subset=['game_over_under', 'total'])

# Categorize sportsbook accuracy
model_game_data['sportsbook_accuracy'] = model_game_data.apply(
    lambda row: 'Overestimated' if row['game_over_under'] > row['total'] 
    else ('Underestimated' if row['game_over_under'] < row['total'] else 'Exact Match'),
    axis=1
)

# Count occurrences
accuracy_counts = model_game_data['sportsbook_accuracy'].value_counts().reset_index()
accuracy_counts.columns = ['Category', 'Count']

# Create bar plot with corrected syntax
plt.figure(figsize=(8, 5))
sns.barplot(data=accuracy_counts, x='Category', y='Count', hue='Category', palette="Blues", legend=False)

plt.xlabel("Sportsbook Prediction Accuracy")
plt.ylabel("Number of Games")
plt.title("How Accurate Are Sportsbook Over/Under Lines?")
plt.show()
````

The initial visualization provides insight into how accurately sportsbooks predict the total points in NFL games. The distribution reveals that sportsbooks most frequently **overestimate** game totals, followed closely by **underestimations**, while exact matches are extremely rare. This pattern suggests that sportsbooks are more likely to set their over/under lines above the actual total points scored, rather than accurately predicting or underestimating them. While minor deviations from perfect accuracy are expected in any predictive model, the consistent tendency toward overestimation raises the question of whether this pattern is the result of **random chance or a deliberate bias in sportsbook line-setting.**

To determine whether sportsbooks systematically misestimate game totals in one direction, we conducted a **chi-square goodness-of-fit test.** This test evaluates whether the observed distribution of overestimated, underestimated, and exact matches significantly deviates from an expected neutral distribution, where each category would occur with similar frequency. The results of this test provide a statistical foundation for understanding whether sportsbooks are **intentionally setting totals higher than expected game outcomes or if this pattern is an artifact of random variation.**

````python
import scipy.stats as stats

# Get observed frequencies
observed_counts = accuracy_counts['Count'].values

# Expected: If sportsbooks were perfect, we assume equal distribution
expected_counts = [sum(observed_counts) / len(observed_counts)] * len(observed_counts)

# Perform chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Display results
print(f"Chi-Square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.6f}")

# Interpretation
if p_value < 0.05:
    print("The differences between sportsbook predictions are statistically significant (p < 0.05).")
else:
    print("No significant difference found in sportsbook predictions (p >= 0.05).")
````

The test results (χ² = 460.33, p < 0.0001) confirm that sportsbooks systematically overestimate total points rather than setting unbiased lines. This is not due to randomness—sportsbooks appear to **intentionally inflate totals**, likely in response to **public betting preferences**. Research shows that **bettors favor the over**, as high-scoring games are perceived as more exciting. By setting lines slightly above expected outcomes, sportsbooks can **exploit this bias**, ensuring more wagers on the over while maintaining their edge. The rarity of exact matches further suggests that sportsbooks prioritize **betting market balance over precise predictions.**

### Parse Weather Data and Aggregate Team Data
- Extract home team's weather features (`temp`, `humidity`, `wind_speed`) from the `weather` column.
- Remove rows with missing values to ensure data integrity.
- Filter the data to only include the last 12 games from each team to maximize fidelity 
- Group the data by team so that we can test our model on games that haven't happened yet later on

````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Parse weather data from `model_game_data`
df = model_game_data.copy()

# Extract weather features
df['temp'] = df['weather'].str.extract(r'Temp: (\d+)°').astype(np.float32)
df['humidity'] = df['weather'].str.extract(r'Humidity: (\d+)%').astype(np.float32)
df['wind_speed'] = df['weather'].str.extract(r'Wind:.* (\d+) mph').astype(np.float32)

# Drop the original 'weather' column
df = df.drop(columns=['weather'])

# Drop rows with missing values

# Ensure data is sorted by team and game date (season and week assumed)
df = df.sort_values(by=['home_team', 'season', 'week'])

# Filter last 12 games for home team
home_data = df.groupby('home_team').tail(8)[[
    'home_team', 'home_total_yards', 'home_rushing_yards', 'home_receiving_yards', 'home_total_epa',
    'home_third_down_%', 'home_pressures', 'home_defense_epa_allowed', 'home_sacks',  'away_sacks', 
    'home_sack_differential', 'home_interceptions', 'home_fumbles_forced', 'home_turnovers', 'home_time_to_throw',
    'home_turnover_differential', 'home_receiving_yards_allowed', 'home_rushing_yards_allowed', 'temp', 'humidity', 'wind_speed'
]].rename(columns=lambda col: col.replace('home_', '')).rename(columns={'home_team': 'team'}).rename(columns={'away_sacks': 'sacks_allowed'})

# Ensure data is sorted by team and game date (season and week assumed)
df = df.sort_values(by=['away_team', 'season', 'week'])

# Filter last 12 games for away team
away_data = df.groupby('away_team').tail(8)[[
    'away_team', 'away_total_yards', 'away_rushing_yards', 'away_receiving_yards', 'away_total_epa', 'away_third_down_%', 'away_pressures', 'away_time_to_throw',
    'away_defense_epa_allowed','away_sacks', 'home_sacks', 'away_sack_differential', 'away_receiving_yards_allowed', 'away_rushing_yards_allowed',
    'away_interceptions', 'away_fumbles_forced', 'away_turnovers', 'away_turnover_differential', 'temp', 'humidity', 'wind_speed', 
]].rename(columns=lambda col: col.replace('away_', '')).rename(columns={'away_team': 'team'}).rename(columns={'home_sacks': 'sacks_allowed'})

# Combine home and away data for later use 
team_data = pd.concat([home_data, away_data], ignore_index=True)
````

### Define Features and Targets
- Define `numeric_features` and `categorical_features` for the dataset.
- Extract features (`X`) and targets (`y_home`, `y_away`) for model training.


````python
# Define numeric and categorical features
numeric_features = [
    'home_total_yards', 'home_total_epa', 'home_turnover_differential', 'home_sack_differential', 
    'home_defense_epa_allowed', 'home_third_down_%', 'away_total_yards', 'away_third_down_%',
    'away_total_epa', 'away_turnover_differential', 'away_sack_differential', 'away_defense_epa_allowed'
]
categorical_features = ['home_team', 'away_team']

# Define target variables: home and away final scores
y_home = df['home_final_score']
y_away = df['away_final_score']
````

### Train-Test Split
- Split the data into training and testing sets with a 75% training and 25% testing ratio.
- Ensure `random_state` for reproducibility.


````python
# Train-test split
X_train, X_test, y_home_train, y_home_test, y_away_train, y_away_test = train_test_split(
    df[numeric_features + categorical_features], y_home, y_away, test_size=0.25, random_state=42
)
````

### Preprocessing
- Normalize numeric features and one-hot encode categorical features using `ColumnTransformer`.
- Apply the transformation to both training and testing data.


````python
# Create the preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),  
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)  
    ],
    remainder='drop'
)

# Preprocess the data
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
````

### Train Models
- Train two separate `RandomForestRegressor` models:
  - One for predicting home team scores.
  - Another for predicting away team scores.


````python
# Train separate models for home and away scores
home_model = RandomForestRegressor(n_estimators=100, random_state=42)
away_model = RandomForestRegressor(n_estimators=100, random_state=42)

home_model.fit(X_train_transformed, y_home_train)
away_model.fit(X_train_transformed, y_away_train)
````

### Evaluate Models
- Use `mean_squared_error` to evaluate the performance of both models on test data.
- Output the Mean Squared Error (MSE) for both home and away predictions as well as their sum


````python
# Make predictions
home_predictions = home_model.predict(X_test_transformed)
away_predictions = away_model.predict(X_test_transformed)

# Compute total points predictions
total_points_predictions = home_predictions + away_predictions

# Compare with sportsbook over/under
df_test = X_test.copy()
df_test['Predicted_Total'] = total_points_predictions
df_test['Actual_Total'] = y_home_test + y_away_test
df_test['Sportsbook_Line'] = df.loc[df_test.index, 'game_over_under']

# Evaluate model performance
home_mse = mean_squared_error(y_home_test, home_predictions)
away_mse = mean_squared_error(y_away_test, away_predictions)
total_mse = mean_squared_error(df_test['Actual_Total'], df_test['Predicted_Total'])

print(f"Home Model MSE: {home_mse:.2f}")
print(f"Away Model MSE: {away_mse:.2f}")
print(f"Total Points Model MSE: {total_mse:.2f}")
````

### Define Prediction Function
- `predict_lines_for_sunday_games`:
  - Takes a list of games (home_team, away_team) and aggregated stats.
  - Predicts the line (points total) as `away_score + home_score`.
  - Returns predictions as a dictionary.


````python
def predict_lines_for_sunday_games(sunday_games, team_data, game_model, preprocessor):
    """
    Predicts the line for all of this Sunday's games using all 12 games per team.

    Parameters:
    - sunday_games (list of tuples): List of (away_team, home_team) for each game.
    - team_data (DataFrame): Data for each team, with one row per game.
    - home_model (RandomForestRegressor): Trained model for predicting home team scores.
    - away_model (RandomForestRegressor): Trained model for predicting away team scores.
    - preprocessor (ColumnTransformer): Preprocessor used for transforming data.

    Returns:
    - predictions (dict): Dictionary with game pairs as keys and predicted lines as values.
    """
    predictions = {}

    for away_team, home_team in sunday_games:
        # Get data for all 12 games of the home and away teams
        home_data = team_data[team_data['team'] == home_team]
        away_data = team_data[team_data['team'] == away_team]

        if home_data.empty or away_data.empty:
            print(f"Data missing for teams: {away_team}, {home_team}")
            continue

        # Create the dataset for all combinations of home and away games
        combined_features = []

        for _, home_row in home_data.iterrows():
            for _, away_row in away_data.iterrows():
                combined_features.append({
                    'home_team': home_team,
                    'away_team': away_team,
                    'home_total_yards': home_row['total_yards'],
                    'home_total_epa': home_row['total_epa'],
                    'home_sacks': home_row['sacks'],
                    'home_interceptions': home_row['interceptions'],
                    'home_fumbles_forced': home_row['fumbles_forced'],
                    'home_rushing_yards': home_row['rushing_yards'],
                    'home_receiving_yards': home_row['receiving_yards'],
                    'home_turnovers': home_row['turnovers'],
                    'home_turnover_differential': home_row['turnover_differential'],
                    'home_sack_differential': home_row['sack_differential'],
                    'home_defense_epa_allowed': home_row['defense_epa_allowed'],
                    'home_receiving_yards_allowed': home_row['receiving_yards_allowed'],
                    'home_rushing_yards_allowed': home_row['rushing_yards_allowed'],
                    'home_third_down_%': home_row['third_down_%'],
                    'home_time_to_throw': home_row['time_to_throw'],
                    'home_pressures': home_row['pressures'],          
                    'temp': home_row['temp'],
                    'humidity': home_row['humidity'],
                    'wind_speed': home_row['wind_speed'],
                    'away_total_yards': away_row['total_yards'],
                    'away_total_epa': away_row['total_epa'],
                    'away_sacks': away_row['sacks'],
                    'away_interceptions': away_row['interceptions'],
                    'away_fumbles_forced': away_row['fumbles_forced'],
                    'away_rushing_yards': away_row['rushing_yards'],
                    'away_receiving_yards': away_row['receiving_yards'],
                    'away_turnovers': away_row['turnovers'],
                    'away_turnover_differential': away_row['turnover_differential'],
                    'away_sack_differential': away_row['sack_differential'],
                    'away_defense_epa_allowed': away_row['defense_epa_allowed'],
                    'away_receiving_yards_allowed': away_row['receiving_yards_allowed'],
                    'away_rushing_yards_allowed': away_row['rushing_yards_allowed'],
                    'away_third_down_%': away_row['third_down_%'],
                    'away_time_to_throw': away_row['time_to_throw'],
                    'away_pressures': away_row['pressures'],
                })

        # Convert combined features to DataFrame
        game_features = pd.DataFrame(combined_features)

        # Preprocess the combined features
        game_features_transformed = preprocessor.transform(game_features)

        # Predict scores for all rows
        game_lines = game_model.predict(game_features_transformed)

        # Calculate lines for all rows
        predicted_lines = game_lines

        # Aggregate the predicted lines (e.g., take the mean)
        predictions[(away_team, home_team)] = predicted_lines.mean()

    return predictions
````
